```python
import pandas as pd
import numpy as np
import seaborn as sns


df = pd.read_excel("data/서울시 시설 데이터/서울지역 대학교 위치.xlsx")
df
```

```python
df.columns = ["학교명","위도","경도"]
df
```

```python
import json
geo_path = 'data/서울시 geojson/seoul.geojson'
geo_str = json.load(open(geo_path, encoding='utf-8'))
```

```python
import folium
seoul_map = folium.Map(location=[37.5617072,126.9745681], zoom_start=12, tiles="cartodbpositron")


    

fmap=folium.Choropleth(geo_data = geo_str,
                       color = "black",
                       nan_fill_color='black',
                       fill_opacity=0.1,
                       fill_color = None).add_to(seoul_map)




for name, lat, lng in zip(df.학교명, df.위도, df.경도):
    folium.CircleMarker([lat, lng],
                        radius=3,         # 원의 반지름
                        fill=True,
                        fill_color='pink',    # 원을 채우는 색
                        fill_opacity=0.7, # 투명도    
                        popup=name
    ).add_to(seoul_map)

    

display(seoul_map)

seoul_map.save('output/seoul_univ.html')
```
